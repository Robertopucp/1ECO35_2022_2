```python
#Pregunta 3
import pyreadr
import matplotlib.pyplot as plt
# Importamos los datos
user = os.getlogin()   # Username
os.chdir(f"C:/Users/{user}/Documents/GitHub/1ECO35_2022_2/Trabajo_grupal/WG8") # Set directorio
base = pyreadr.read_r(r"../../data/wage2015_subsample_inference.RData")
data = base["data"]

# Cambiamos los valores para que cada nivel de educación tenga un valor único
data.hsg.replace((1, 0), (2, 0), inplace=True)
data.scl.replace((1, 0), (3, 0), inplace=True)
data.clg.replace((1, 0), (4, 0), inplace=True)
data.ad.replace((1, 0), (5, 0), inplace=True)

# Creamos una columna de ceros
data['edu']=0

# Creamos una sola variable "edu" con los distintos niveles de educación

data['edu'] = data["shs"] + data["hsg"]+ data["scl"]+ data["clg"]+ data["ad"]

# Cantidad por tipo de educación
base = data.groupby([ 'edu' ])['edu'].size()

# Labels para añadir etiquetas
labels=['Secundaria incompleta','Secundaria completa' , 'Universitaria incompleta', 'Universitaria completa', 'Grado educativo avanzado']


#Creamos el gráfico
plt.figure(figsize=(10,10))
ax = plt.pie(base, labels=labels, autopct='%.1f %%')
plt.title('Porcentaje de personas según nivel educativo')
plt.show
```
