```python
#Se carga la base de datos del modulo 2
enaho02 = pd.read_stata("C:/Users/egus2/Desktop/CURSOS OCTAVO CICLO/R Y PYTHON\WG6/737-Modulo02/Enaho01-2020-200.dta", convert_categoricals = False)
enaho02
```

```python
#Se carga la base de datos del modulo 34
enaho34 = pd.read_stata("C:/Users/egus2/Desktop/CURSOS OCTAVO CICLO/R Y PYTHON\WG6/737-Modulo34/sumaria-2020.dta", convert_categoricals = False)
enaho34
```

```python
#Se extraen solo las columnas que sirven para identificar hogares así como el p208a que sirve para saber a edad de cada individuo
enaho02 = enaho02[['conglome','vivienda','hogar','p208a']]
enaho02
```

```python
#Se realiza la agrupación de los invidividuos por conglomerado, vivienda y hogar.
#Adicionalmente se indica que se extraiga el máximo valor de edad por hogar para saber si algún miembro es apto para el programa pensión 65
enaho02 = enaho02.groupby( ["conglome", "vivienda", "hogar" ], as_index = False)[['p208a']].max()
enaho02
```

```python
#Se seleccionan las columnas de interés del modulo34 para poder hacer merge con el modulo02
#También se incluye a la columna pobreza para saber si el hogar califica para el programa
enaho34 = enaho34[["conglome", "vivienda", "hogar",'pobreza']]
enaho34.sort_values(by = 'conglome')#Se ordena por conglomerado para ver que los registros corresponden a los del modulo02
```

```python
#Se hace merge a los dos módulos, se tiene como base al modulo02
enaho_merge = pd.merge(enaho02, enaho34)
enaho_merge
```

```python
#Se crea la variable dummy de acuerdo a los criterios:
#Al menos un miembro mayor de 65 años
#Que el hogar esté considerado como pobre (pobre extremo y pobre no extremo)
#Se indica que la dummy tenga un valor de 1 cuando se cumple con los criterios para el programa
#y se tendrá un valor de 0 si no se cumplen los criterios. 
enaho_merge["dummy"] = np.where((enaho_merge["p208a"] > 65) & (enaho_merge["pobreza"] < 3), "1", "0")
enaho_merge
```

```python
#Para comprobar la dummy se realiza un conteo de los registros que cumplen y no cumplen con los requisitos.
enaho_merge[["dummy"]].value_counts()
```
