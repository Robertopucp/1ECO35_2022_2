## 3. Keywords especiales (*args and *kwargs) 
Construya una funciÃ³n que permite realizar las siguientes operaciones a un vector: estandarizar y reescalar (funciÃ³n del ejercicio 2). Use **kwargs para definir estas operaciones y *args para el vector.

```python
# Creamos la funciÃ³n "calculator" que estandarizarÃ¡ y reescalara un vector

def calculator( *args, **kwargs):
    
    vector = np.array( list(args) )  # transforma *args a lista, luego la transforma en un array
    
    if ( kwargs[ 'function' ] == "estandarizar" ) :
        # Get the first value
        def reescalar (x, mean, std):
            out = (x - mean) / std
            return out
        
        result = list( map( lambda x, mean = np.mean(vector), std = np.std(vector): reescalar(x, mean, std), vector)  )
    
    elif ( kwargs[ 'function' ] == "reescalar" ) :
        # Get the second value
        def reescalar (x, min, max):
            out = (x - min) / (max - min)
            return out
        
        result = list( map( lambda x, min = np.min(vector), max = np.max(vector): reescalar(x, min, max), vector)  ) 
        
    else:
        raise ValueError( f"The function argument {kwargs[ 'function' ]} is not supported." )
        # Mensaje de error por tipo de argumento

    return result
```

```python
calculator(np.arange(100), function='estandarizar')
```

```python
calculator(np.arange(100), function='reescalar')
```
