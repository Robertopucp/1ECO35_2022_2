#### [Agg](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.agg.html)
Aggregate using one or more operations over the specified axis.


```python
df_urban_no_dpl.groupby( [ 'CCDD' ,'CCPP' , 'CCDI' ], as_index = False ).agg( { "P172D": "mean" } )
```

```python
df3_rec = df_urban_no_dpl.groupby( [ 'CCDD' ,'CCPP' , 'CCDI' ] , as_index = False ).agg( recycle_median = ( 'P172D', np.median ), recycle_mean = ( 'P172D', np.mean ) )
df3_rec
```

```python

# Aplicar esta sentencia al crear sus variables pues P107A presenta missing values. De esta forma
# se crea variables dummies con presencia de missing values 

new_data['Dummy_1'] = np.where(new_data['P107A'] == "Contrato de Compra-Venta", 1, np.where(new_data['P107A'].notnull(), 0, np.nan))

# Cuando es falso la sentencia no coloca 0 sino que colocará cero si el valor es diferente a un missing value 
```

```python
new_data['Dummy_1'].unique()
```

# Examples

```python
new_data['ESTRATO'].unique()
```

```python
# Check missing values

new_data['ESTRATO'].isna().any()
```

```python
# create a dummy for estrato D 

new_data['Dummy_2'] = np.where(new_data['ESTRATO'] == "Estrato D", 1, 0)

# Estructura  where:  np.where( condition, verdadero: coloca 1, falso: coloca 0)
```

```python
# Create dummy base on multiples values from other variable 

new_data['Dummy_3'] = np.where(new_data['ESTRATO'].isin(["Estrato D", "Estrato E más bajo", "Estrato C"]), 1, 0)
```

```python
df = new_data.groupby( [ 'CCDD' ,'CCPP' , 'CCDI' ] , as_index = False ).agg( estrato_D_median = ( 'Dummy_2', np.median ), 
                        estrato_D_mean = ( 'Dummy_2', np.mean ), estrato_D_max = ( 'Dummy_2', np.max ) , estrato_D_min = ( 'Dummy_2', np.min ) ,
                        estrato_D_total = ( 'Dummy_2', np.sum ) , total_hogares_distrito = ('HOGAR', np.size))

# estrato_D_median : la mediana de Dummy_2 por distrito 
# estrato_D_median : la media de Dummy_2 por distrito (porcentaje de hogares que pertenecen al estrato D a nivel distrital)
# estrato_D_max: el maximo valor de la variable Dummy_2 por distrito 
# estrato_D_min: el minimo valor de la variable Dummy_2 por distrito 
# estrato_D_total: total de hogares que pertenecen al estrato D 
# estrato_D_total: total de hogares por distrito (se está contando la cantidad de hogares por distrito)
```

```python
# save in spss format 

import pyreadstat

pyreadstat.write_sav(df,  r"../../data/enapres_2020_ch_100/736-Modulo1618/test1.sav")
```
