```python
import numpy as np
import pandas as pd
```

```python
#Importamos los archivos de los Estadísticos del programa Juntos
df1 = pd.read_html('2014.xls')
df2 = pd.read_html('2015.xls')
df3 = pd.read_html('2016.xls')
df4 = pd.read_html('2017.xls')
df5 = pd.read_html('2018.xls')
df6 = pd.read_html('2019.xls')
df7 = pd.read_html('2020.xls')
df8 = pd.read_html('2021.xls')
df9 = pd.read_html('2022.xls')
```

```python
# Revisamos la estructura de los datos que presentan
df1[1]
```

```python
#Filtramos y nos quedamos con los datos requeridos
df1 = df1[1][3:-1][[0,7]].drop([11, 12],axis=0).T
df2 = df2[1][3:-1][[7]].drop([11, 12],axis=0).T
df3 = df3[1][3:-1][[7]].drop([11, 12],axis=0).T
df4 = df4[1][3:-1][[7]].drop([11, 12],axis=0).T
df5 = df5[1][3:-1][[7]].drop([11, 12],axis=0).T
df6 = df6[1][3:-1][[7]].drop([11, 12],axis=0).T 
df7 = df7[1][3:-1][[7]].drop([11, 12],axis=0).T
df8 = df8[1][3:-1][[7]].drop([11, 12],axis=0).T
df9 = df9[1][3:-1][[7]].drop([11, 12],axis=0).T
```

```python
df1
```

```python
#Concatenamos los datos de cada año en un solo dataframe
lista = [df1,df2,df3,df4,df5,df6,df7,df8,df9]
df_concat = pd.concat(lista)
df_concat.reset_index(drop = True,inplace = True)
df_concat
```

```python
#Añadimos la columna de años al dataframe
años = pd.DataFrame(['years', 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]).reset_index(drop = True)
df = pd.concat([df_concat,años], axis=1)
```

```python
#Ponemos en formato nuestra tabla para presentarlo de manera mas clara
df = df.rename(columns=df.iloc[0]).drop(df.index[0]).style.hide_index()
df
```
